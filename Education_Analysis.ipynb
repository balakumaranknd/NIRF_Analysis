# Male : Female Ratio Analysis

```python
m_iitm = st_dt_iitm.iloc[0:4,0:1]
m_iitd = st_dt_iitd.iloc[0:3,0:1]
m_iitb = st_dt_iitb.iloc[0:3,0:1]
m_iitk = st_dt_iitk.iloc[0:2,0:1]
m_iitkh = st_dt_iitkh.iloc[0:5,0:1]
```

```python
m_iitm
```

```python
temp = pd.merge(left=m_iitm, right=m_iitd, how='inner',suffixes=('Madras','Delhi'),left_index=True,right_index=True)
temp = pd.merge(left=temp, right=m_iitb, how='inner',suffixes=['Bombay'],left_index=True,right_index=True)
temp.rename(mapper = {'No. of Male\nStudentsMadras' : 'Madras','No. of Male\nStudentsDelhi' : 'Delhi',
            'No. of Male\nStudents' : 'Bombay'},axis=1,inplace=True)
temp = pd.merge(left=temp, right=m_iitk, how='inner',suffixes=(None,'Kanpur'),left_index=True,right_index=True)
temp.rename(mapper = {'No. of Male\nStudents' : 'Kanpur'},axis=1,inplace=True)
males = pd.merge(left=temp, right=m_iitkh, how='inner',suffixes=['Kharagpur'],left_index=True,right_index=True)
males.rename(mapper = {'No. of Male\nStudents' : 'Kharagpur'},axis=1,inplace=True)
```

```python
f_iitm = st_dt_iitm.iloc[0:4,1:2]
f_iitd = st_dt_iitd.iloc[0:3,1:2]
f_iitb = st_dt_iitb.iloc[0:3,1:2]
f_iitk = st_dt_iitk.iloc[0:2,1:2]
f_iitkh = st_dt_iitkh.iloc[0:5,1:2]
```

```python
temp = pd.merge(left=f_iitm, right=f_iitd, how='inner',suffixes=('Madras','Delhi'),left_index=True,right_index=True)
temp = pd.merge(left=temp, right=f_iitb, how='inner',suffixes=['Bombay'],left_index=True,right_index=True)
temp.rename(mapper = {'No. of Female\nStudentsMadras' : 'Madras','No. of Female\nStudentsDelhi' : 'Delhi',
            'No. of Female\nStudents' : 'Bombay'},axis=1,inplace=True)
temp = pd.merge(left=temp, right=f_iitk, how='inner',suffixes=(None,'Kanpur'),left_index=True,right_index=True)
temp.rename(mapper = {'No. of Female\nStudents' : 'Kanpur'},axis=1,inplace=True)
females = pd.merge(left=temp, right=f_iitkh, how='inner',suffixes=['Kharagpur'],left_index=True,right_index=True)
females.rename(mapper = {'No. of Female\nStudents' : 'Kharagpur'},axis=1,inplace=True)
```

```python
females
```

```python
males.reset_index(level=0,inplace=True)
females.reset_index(level=0,inplace=True)
```

```python
males_t = males.T
males_t.rename(columns={0 : 'UG 4 years',1 : 'PG 2 years'},inplace=True)
females_t = females.T
females_t.rename(columns={0 : 'UG 4 years',1 : 'PG 2 years'},inplace=True)
```

```python
males_t.reset_index(level=0,inplace=True)
females_t.reset_index(level=0,inplace=True)
```

```python
males_t.rename(columns={'index' : 'City'},inplace=True)
females_t.rename(columns={'index' : 'City'},inplace=True)
males_t.drop(index=0,axis=0,inplace=True)
females_t.drop(index=0,axis=0,inplace=True)
```

```python
females_t
```

```python
dfs = {'Males' : males_t,'Females' : females_t}
```

```python
writer = pd.ExcelWriter('MaleFemaleAnalysis.xlsx',engine='xlsxwriter')

for sheet_name in dfs.keys():
    dfs[sheet_name].to_excel(writer, sheet_name=sheet_name,index=False)
writer.save()
```
