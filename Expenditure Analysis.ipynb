```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.float_format', lambda x: '%10.10f' % x)
```

```python
capex_iitm = pd.read_excel(open('Capex.xlsx','rb'),sheet_name="1_iitm")
capex_iitd = pd.read_excel(open('Capex.xlsx','rb'),sheet_name="2_iitd")
capex_iitb = pd.read_excel(open('Capex.xlsx','rb'),sheet_name="3_iitb")
capex_iitk = pd.read_excel(open('Capex.xlsx','rb'),sheet_name="4_iitk")
capex_iitkh = pd.read_excel(open('Capex.xlsx','rb'),sheet_name="5_iitkh")
```

```python
opex_iitm = pd.read_excel(open('Opex.xlsx','rb'),sheet_name="1_iitm")
opex_iitd = pd.read_excel(open('Opex.xlsx','rb'),sheet_name="2_iitd")
opex_iitb = pd.read_excel(open('Opex.xlsx','rb'),sheet_name="3_iitb")
opex_iitk = pd.read_excel(open('Opex.xlsx','rb'),sheet_name="4_iitk")
opex_iitkh = pd.read_excel(open('Opex.xlsx','rb'),sheet_name="5_iitkh")
```

```python
capital = [capex_iitm,capex_iitd,capex_iitb,capex_iitk,capex_iitkh]
operational = [opex_iitm,opex_iitd,opex_iitb,opex_iitk,opex_iitkh]


for dataframe in capital:
    dataframe.rename(columns={"Unnamed: 0" : 'Category'},inplace=True)
    
    dataframe.set_index('Category',inplace=True)

for dataframe in operational:
    dataframe.rename(columns={"Unnamed: 0" : 'Category'},inplace=True)
    
    dataframe.set_index('Category',inplace=True)
```

### 1. IIT Madras

```python
capex_iitm_t = capex_iitm.T
capex_iitm_t['Total'] = capex_iitm_t['Library ( Books, Journals and e-Resources only)'] + \
                        capex_iitm_t['New Equipment and software for Laboratories'] + \
                        capex_iitm_t['Other expenditure on creation of Capital Assets (For setting up\nclassrooms, seminar hall, conference hall , library, Lab, Engg\nworkshops excluding expenditure on Land and Building)']
capex_iitm_t
```

```python

capex_iitm.T.plot(kind='bar')
plt.title('Capital Expenditure - IIT Madras')
# after plotting the data, format the labels
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])
plt.gcf().set_size_inches(10, 10)

```

```python
opex_iitm_t = opex_iitm.T
opex_iitm_t['Total'] = opex_t['Salaries (Teaching and Non Teaching staff)'] + \
                    opex_t['Maintenance of Academic Infrastructure or consumables and\nother running expenditures(excluding maintenance of hostels\nand allied services,rent of the building, depreciation cost, etc)'] + \
                    opex_t['Seminars/Conferences/Workshops']
        
opex_iitm_t
```

```python

opex_iitm.T.plot(kind='bar')
plt.title('Operational Expenditure - IIT Madras')
# after plotting the data, format the labels
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])
plt.gcf().set_size_inches(10, 10)
plt.show()
```
